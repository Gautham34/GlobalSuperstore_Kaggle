# Superstore Sales

```python
#importing libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style('whitegrid')
```

```python
#Reading the csv file
df = pd.read_csv('superstore.csv')
```

```python
#Getting the whole info of the dataset

df.info()
```

```python
#First 50 entries

df.head(50)
```

```python
df.describe()
```

```python
#Checking for null values

df.isnull().sum()
```

```python
#Checking duplicate values

df.duplicated().sum()
```

```python
#Changing date format to datetime

df['Order.Date'] = pd.to_datetime(df['Order.Date'], errors='coerce')
df['Ship.Date'] = pd.to_datetime(df['Ship.Date'], errors='coerce')
```

```python
df[['Order.Date','Ship.Date']].head()
```

```python
#Creating seperate columns for year, month and quarter

df['Year'] = df['Order.Date'].dt.year
df['Month'] = df['Order.Date'].dt.month
df['Quarter'] = df['Order.Date'].dt.quarter
```

```python
#Amount of days taken for delivery

df['Delivery.Days'] = (df['Ship.Date'] - df['Order.Date']).dt.days
```

```python
#Checking profit margin

df['Profit.Margin'] = df['Profit'] / df['Sales']
df.head()
```

```python
high_margin = df[df['Profit.Margin'] > 0.5]
high_margin
```

```python
df['Profit.Margin.%'] = (df['Profit'] / df['Sales']) * 100
df.head()
```

```python
df.head()
```

```python
#Visualisation of monthly trends of sales

monthly_sales = df.set_index('Order.Date').resample('ME')['Sales'].sum()

plt.figure(figsize=(10,4))
monthly_sales.plot(color='red')
plt.title("Monthly Sales Trend")
plt.ylabel("Sales")
plt.show()
```

```python
#Visualisation of sales of each category

plt.figure(figsize=(10,4))
df.groupby('Category')['Sales'].sum().plot(kind='bar', color='skyblue')
plt.title("Sales by Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Visualisation of monthly trends of sales

top_states = df.groupby('State')['Profit'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,4))
top_states.plot(kind='bar', color='green')
plt.title("Top 10 States by Profit")
plt.ylabel("Profit")
plt.show()
```

```python
#Pie-chart showing share of each category

category_sales = df.groupby('Category')['Sales'].sum()

plt.figure(figsize=(6,6))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=90)
plt.title("Sales Share by Category")
plt.show()
```

```python
#Stacked bar chart showing sales of each segment per year

segment_year = df.groupby(['Year','Segment'])['Sales'].sum().unstack()

segment_year.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title("Sales by Segment Over Years")
plt.ylabel("Sales")
plt.show()
```

```python
#Heatmap showing sales in each month of a year

pivot = df.pivot_table(index='Year', columns='Month', values='Sales', aggfunc='sum')

plt.figure(figsize=(10,6))
sns.heatmap(pivot, cmap="YlGnBu", annot=True, fmt=".0f")
plt.title("Sales Heatmap (Year vs Month)")
plt.show()
```
